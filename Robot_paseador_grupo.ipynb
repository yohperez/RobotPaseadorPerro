```python
import time

class RobotPaseador:
    def __init__(self, modelo):
        self.modelo = modelo
        self.bateria = 45  # Simulación de estado inicial
        self.ubicacion = "Central"
        self.deposito_bolsas = 0
        self.normas_madrid = []

    def configurar_sistema(self):
        print(f"\n--- [⚙️ Configuración {self.modelo}] ---")

        # Lógica de batería
        if self.bateria < 50:
            print(f"Batería al {self.bateria}%. Cargando hasta el 100%...")
            self.bateria = 100
        else:
            print("Nivel de batería óptimo para iniciar.")

        print("Descargando normativa de Madrid: Ordenanza de Tenencia de Animales...")
        self.normas_madrid = ["Limpiar orina con agua", "Recoger excrementos", "Uso de correa"]
        print("Rutas sincronizadas. Elementos de limpieza verificados.")

    def reaccionar_a_entorno(self, evento):
        if evento == "charco":
            print("[SISTEMA]: Charco detectado. Recalculando paso para evitarlo.")
        elif evento == "perro_defeca":
            print("[SISTEMA]: Protocolo de limpieza iniciado.")
            print("-> Sacando dispensador, tomando bolsa, inclinándose, cerrando y guardando bolsa.")
            self.deposito_bolsas += 1
        elif evento == "lluvia":
            print("[SISTEMA]: Llueve. Buscando refugio seco para el perro...")
            time.sleep(2)
            print("[SISTEMA]: Lluvia finalizada. Reanudando ruta.")
        elif evento == "pis":
            print("[SISTEMA]: Limpiando orina con la botella de agua (Normativa Madrid).")

    def caminar(self):
        print("-> Caminando por la vereda. Seguridad: Calle evitada.")

class AppPaseo:
    def __init__(self):
        self.planes = {
            "Bronce": {"min": 15, "mtrs": 350, "precio": 15},
            "Plata": {"min": 30, "mtrs": 750, "precio": 30},
            "Oro": {"min": 60, "mtrs": 1000, "precio": 45}
        }

    def gestionar_reserva(self):
        print("\n--- [📱 App BotPro: Registro] ---")
        email = input("Email: ")
        password = input("Contraseña: ")
        nombre_dueño = input("Nombre del dueño: ")
        apellido = input("Apellido: ")
        dni = input("DNI: ")
        input("¿Acepta términos de protección de datos? (S/N): ")

        print("\n--- [📱 Reserva de Paseo] ---")
        mascota = input("Nombre de la mascota: ")
        chip = input("Código Microchip: ")
        fecha = input("Fecha del paseo: ")
        hora = input("Hora de recogida: ")
        tamano = input("Tipo de perro (Grande/Mediano/Pequeño): ")

        print("Planes disponibles: Bronce (15€), Plata (30€), Oro (45€)")
        plan_sel = input("Seleccione plan: ").capitalize()

        total = self.planes[plan_sel]["precio"]

        extra = input("¿Deseas tiempo extra? (0.50€/min) (si/no): ").lower()
        if extra == "si":
            minutos = int(input("Ingresa minutos extras: "))
            total += (minutos * 0.50)

        indicaciones = input("Indicaciones especiales (dejar en blanco si no hay): ")
        domicilio = input("Introduce datos de domicilio: ")

        return {
            "dueño": nombre_dueño,
            "mascota": mascota,
            "total": total,
            "domicilio": domicilio,
            "codigo_seguridad": "7744"
        }

# --- EJECUCIÓN DEL PROCESO ---

# 1. Reserva y Configuración
app = AppPaseo()
reserva = app.gestionar_reserva()

robot = RobotPaseador("BotPro v2.1")
robot.configurar_sistema()

# 2. Inicio del Servicio
print(f"\n--- [🚀 Inicio del Servicio] ---")
print(f"Robot llega a {reserva['domicilio']}. *Toca el timbre*")
print(f"Robot: 'Hola, {reserva['dueño']}, he venido por {reserva['mascota']}'")

cod_ingresado = input(f"Humano, ingresa el código mostrado en el robot ({reserva['codigo_seguridad']}): ")
if cod_ingresado == reserva['codigo_seguridad']:
    print("Arnés desbloqueado. Correa asegurada.")
    input("Humano: Confirma inicio de paseo en la App (Presiona Enter)")
else:
    print("Código incorrecto. Por seguridad el servicio se cancela.")
    exit()

# 3. El Paseo
print("\n--- [🐕 En el Paseo] ---")
robot.caminar()
robot.reaccionar_a_entorno("charco")
robot.reaccionar_a_entorno("perro_defeca")
robot.reaccionar_a_entorno("pis")
robot.reaccionar_a_entorno("lluvia")
print("Robot: 'Paseo completado según tiempo y distancia contratada'.")

# 4. Devolución y Retorno
print("\n--- [🏠 Devolución] ---")
print(f"Entregando a {reserva['mascota']}. El humano retira el arnés.")
input("Deslice el botón en la app para confirmar recepción (Enter)")
print("Robot: '¡Adiós!'. Recalculando posición para volver a la central...")

# 5. Post-Servicio Robot
print("\n--- [🔧 Mantenimiento Central] ---")
print(f"Vaciando depósito: {robot.deposito_bolsas} desechos eliminados.")
robot.bateria = 100
print("Robot cargado y listo para el siguiente servicio.")

# 6. Post-Servicio Humano (Pago y Feedback)
print("\n--- [💳 Pago y Calificación] ---")
print(f"Total a pagar: {reserva['total']}€")
metodo_pago = input("Seleccione método de pago: ")
propina = input("Seleccionar propina: ")
print("Confirmando pago... ¡Éxito!")

# Calificación
satis = input("Del 1 al 10 ¿Qué tan satisfecho estás con el servicio?: ")
puntual = input("Calificar puntualidad del 1 al 10: ")
sugiere = input("¿Desea agregar sugerencias? (si/no): ").lower()
if sugiere == "si":
    input("Ingrese su comentario: ")

print(f"\n{reserva['dueño']}, gracias por confiar en BotPro, recuerda recomendar nuestros servicios. Ten un buen día.")
```

```python
import math
import time

class RobotPaseadorMadrid:
    def __init__(self, nombre_robot, bateria, num_bolsas=8):
        # --- Configuración del Robot ---
        self.nombre_robot = nombre_robot
        self.bateria = bateria
        self.num_bolsas = num_bolsas
        self.capacidad_max_bolsas = 10
        self.deposito_desechos = 0

        # --- Datos del Paseo (se llenan con la entrevista) ---
        self.nombre_perro = ""
        self.destino = ""
        self.origen = "Centro de Deportes de La Cebada (La Latina)"

        # --- Coordenadas y Mapas (Madrid) ---
        self.pos_actual_gps = (40.4110, -3.7092) # La Cebada
        self.mapa_destinos = {
            "Retiro": (40.4153, -3.6835),
            "Plaza Mayor": (40.4153, -3.7073),
            "Madrid Rio": (40.4121, -3.7210)
        }

        # --- Equipamiento de Seguridad ---
        self.equipamiento = {
            "correa_corta": True,
            "chip_identificativo": True,
            "seguro_rc": True,
            "botella_agua_limpieza": True
        }

    def saludar_y_entrevistar(self):
        """Punto 1: Interacción Inicial"""
        print(f"\n🤖 [ROBOT]: ¡Hola! Soy {self.nombre_robot}, su asistente de paseo.")
        self.nombre_perro = input("👤 [DUEÑO]: ¿Cómo se llama el perro hoy? -> ")

        print(f"🤖 [ROBOT]: ¡Qué nombre tan genial! Hola, {self.nombre_perro}.")
        print(f"🤖 [ROBOT]: ¿A dónde quieres que vayamos? (Opciones: {', '.join(self.mapa_destinos.keys())})")
        eleccion = input("👤 [DUEÑO]: Indique el destino -> ").title().strip()

        if eleccion in self.mapa_destinos:
            self.destino = eleccion
        else:
            self.destino = "Ruta Libre"
            print(f"🤖 [ROBOT]: No tengo el mapa de '{eleccion}', pero exploraremos con seguridad.")

    def verificar_preparacion(self):
        """Paso 2: Seguridad y Batería"""
        print(f"\n--- 🛡️ Verificando Protocolos para {self.nombre_robot} ---")

        # Lógica de carga automática
        if self.bateria < 50:
            print(f"⚠️ Batería al {self.bateria}%. Realizando carga rápida en La Cebada...")
            self.bateria = 100

        # Verificación de elementos físicos
        faltantes = [item for item, estado in self.equipamiento.items() if not estado]
        if faltantes or self.num_bolsas < 1:
            print(f"❌ ERROR: Faltan elementos: {faltantes if faltantes else 'Bolsas'}.")
            return False

        print("✅ Equipo completo y batería cargada.")
        return True

    def identificar_normas_transito(self):
        """Paso 3: Normativa DGT y Madrid"""
        print("\n--- 🚦 Normas de Tránsito Madrid ---")
        normas = [
            "1. Caminar siempre por la vereda lejos de la calzada.",
            "2. Cruzar solo en pasos de cebra con semáforo verde.",
            f"3. {self.nombre_perro} debe ir con correa corta por seguridad vial.",
            "4. Diluir orina con la botella de agua (Ordenanza Municipal)."
        ]
        for norma in normas:
            print(f"   [REGLA] {norma}")

    def planificar_ruta(self):
        """Paso 4: Cálculo Matemático de Distancia"""
        if self.destino in self.mapa_destinos:
            x1, y1 = self.pos_actual_gps
            x2, y2 = self.mapa_destinos[self.destino]
            # Distancia Euclidiana: sqrt((x2-x1)^2 + (y2-y1)^2)
            distancia = math.sqrt((x2 - x1)**2 + (y2 - y1)**2) * 111 # Aprox km
            print(f"\n📍 Ruta: {self.origen} ➔ {self.destino}.")
            print(f"📏 Distancia estimada: {distancia:.2f} km.")
        else:
            print(f"\n📍 Ruta: Paseo libre por La Latina.")

    def ejecutar_paseo(self):
        """Paso 5: Acción y Eventos"""
        print(f"\n--- 🐕 Paseando a {self.nombre_perro} ---")
        print("🚶 Caminando por la acera de Calle Colegiata...")
        print("🛑 Semáforo rojo en Tirso de Molina. Esperando...")
        time.sleep(1)
        print("🟢 Verde. Cruzando con cuidado.")

        # Simulación de obstáculos y necesidades
        self.detectar_obstaculos("Charco en Calle Huertas")
        self.gestionar_desechos(certeza_ia=0.95)

        print(f"🏁 ¡Llegamos a {self.destino}! {self.nombre_perro} está muy contento.")

    def detectar_obstaculos(self, tipo):
        if "Charco" in tipo:
            print(f"🌊 {tipo} detectado. Esquivando por la zona seca.")

    def gestionar_desechos(self, certeza_ia):
        if certeza_ia > 0.8:
            print(f"💩 {self.nombre_perro} ha defecado. Iniciando protocolo de limpieza...")
            if self.num_bolsas > 0:
                print("   [Acción]: Recogiendo con bolsa y aplicando agua de la botella.")
                self.num_bolsas -= 1
                self.deposito_desechos += 1
            else:
                print("   ⚠️ ¡ALERTA! Sin bolsas. Buscando dispensador público.")

    def protocolo_cierre(self):
        """Paso 6: Mantenimiento Final"""
        print("\n--- 🏠 Finalizando Servicio ---")
        print(f"1. {self.nombre_perro} entregado a su dueño.")
        print(f"2. Vaciando {self.deposito_desechos} bolsas en contenedor orgánico.")
        print(f"3. Cargando batería al 100%. Bolsas restantes: {self.num_bolsas}.")
        print(f"🤖 [ROBOT]: ¡Gracias por confiar en {self.nombre_robot}! Hasta pronto.")

# --- INICIO DEL PROGRAMA ---

# Instanciamos el robot
mi_bot = RobotPaseadorMadrid("Chulapo-Bot", bateria=45)

# Ejecutamos la secuencia lógica
mi_bot.saludar_y_entrevistar()

if mi_bot.verificar_preparacion():
    mi_bot.identificar_normas_transito()
    mi_bot.planificar_ruta()
    mi_bot.ejecutar_paseo()
    mi_bot.protocolo_cierre()
```

```python
import time

class RobotPaseador:
    def __init__(self, modelo):
        self.modelo = modelo
        self.bateria = 45  # Simulación de estado inicial
        self.ubicacion = "Central"
        self.deposito_bolsas = 0
        self.normas_madrid = []

    def configurar_sistema(self):
        print(f"\n--- [⚙️ Configuración {self.modelo}] ---")

        # Lógica de batería
        if self.bateria < 50:
            print(f"¡Hola! Mi batería está al {self.bateria}%. Me cargaré un poquito hasta el 100% para dar el mejor paseo.")
            self.bateria = 100
        else:
            print("¡Energía a tope! Estoy listo para comenzar esta aventura.")

        print("Sincronizando la normativa de Madrid con mucho cuidado...")
        self.normas_madrid = ["Limpiar orina con agua", "Recoger excrementos", "Uso de correa"]
        print("Rutas listas y equipo de limpieza preparado. ¡Todo en orden!")

    def hidratar_mascota(self, mascota, minuto):
        print(f"\n[💧 HIDRATACIÓN]: Han pasado {minuto} minutos. Me detengo suavemente para que {mascota} beba agua fresca.")
        print(f"[📸 APP NOTIFICACIÓN]: Enviando foto de {mascota} bebiendo a la App del dueño... ¡Se ve muy feliz!")

    def reaccionar_a_entorno(self, evento):
        if evento == "charco":
            print("[SISTEMA]: ¡Uy, un charco! Recalculando nuestra ruta para que no se moje las patitas.")
        elif evento == "perro_defeca":
            print("[SISTEMA]: ¡Oh! Hora de cumplir con mi deber de limpieza.")
            print("-> Con mucho cuidado: saco bolsa, recojo, cierro y guardo para mantener limpia la ciudad.")
            self.deposito_bolsas += 1
        elif evento == "lluvia":
            print("[SISTEMA]: ¡Están cayendo gotas! Buscando un refugio seco y seguro de inmediato.")
            time.sleep(2)
            print("[SISTEMA]: Ya paró de llover. ¡Podemos seguir disfrutando del aire libre!")
        elif evento == "pis":
            print("[SISTEMA]: Aplicando agua con jabón según la normativa de Madrid. ¡Impecable!")

    def caminar(self):
        print("-> Caminando alegremente por la acera. Mantengo a tu mascota lejos del tráfico por seguridad.")

class AppPaseo:
    def __init__(self):
        self.planes = {
            "Bronce": {"min": 15, "mtrs": 350, "precio": 15},
            "Plata": {"min": 30, "mtrs": 750, "precio": 30},
            "Oro": {"min": 60, "mtrs": 1000, "precio": 45}
        }

    def gestionar_reserva(self):
        print("\n--- [📱 App BotPro: Registro] ---")
        email = input("Email: ")
        password = input("Contraseña: ")
        nombre_dueño = input("Nombre del dueño: ")
        apellido = input("Apellido: ")
        dni = input("DNI: ")
        input("¿Acepta términos de protección de datos? (S/N): ")

        print("\n--- [📱 Reserva de Paseo] ---")
        mascota = input("Nombre de la mascota: ")
        chip = input("Código Microchip: ")
        fecha = input("Fecha del paseo: ")
        hora = input("Hora de recogida: ")
        tamano = input("Tipo de perro (Grande/Mediano/Pequeño): ")

        print("Planes disponibles: Bronce (15€), Plata (30€), Oro (45€)")
        plan_sel = input("Seleccione plan: ").capitalize()

        total = self.planes[plan_sel]["precio"]
        minutos_totales = self.planes[plan_sel]["min"]

        extra = input("¿Deseas tiempo extra? (0.50€/min) (si/no): ").lower()
        if extra == "si":
            minutos_extra = int(input("Ingresa minutos extras: "))
            total += (minutos_extra * 0.50)
            minutos_totales += minutos_extra

        indicaciones = input("Indicaciones especiales (dejar en blanco si no hay): ")
        domicilio = input("Introduce datos de domicilio: ")

        return {
            "dueño": nombre_dueño,
            "mascota": mascota,
            "total": total,
            "domicilio": domicilio,
            "minutos": minutos_totales,
            "codigo_recogida": "7744",
            "codigo_entrega": "1234" # Código de seguridad para la llegada
        }

# --- EJECUCIÓN DEL PROCESO ---

# 1. Reserva y Configuración
app = AppPaseo()
reserva = app.gestionar_reserva()

robot = RobotPaseador("BotPro v2.1")
robot.configurar_sistema()

# 2. Inicio del Servicio
print(f"\n--- [🚀 Inicio del Servicio] ---")
print(f"Robot llega a {reserva['domicilio']}. *Toca el timbre con una melodía amigable*")
print(f"Robot: '¡Hola, {reserva['dueño']}! Qué alegría verte. He venido para llevar a {reserva['mascota']} a divertirse.'")

cod_ingresado = input(f"Humano, por favor ingresa el código de seguridad ({reserva['codigo_recogida']}): ")
if cod_ingresado == reserva['codigo_recogida']:
    print("¡Perfecto! Arnés desbloqueado y ajustado con suavidad. ¡Estamos listos!")
    input("Humano: Confirma inicio de paseo en la App (Presiona Enter)")
else:
    print("Vaya, el código no coincide. Por seguridad de la mascota, no puedo iniciar el servicio.")
    exit()

# 3. El Paseo con Hidratación
print("\n--- [🐕 En el Paseo] ---")
robot.caminar()
robot.reaccionar_a_entorno("charco")

# Simulación de hidratación cada 10 minutos
for m in range(10, reserva['minutos'] + 1, 10):
    robot.hidratar_mascota(reserva['mascota'], m)

robot.reaccionar_a_entorno("perro_defeca")
robot.reaccionar_a_entorno("pis")
robot.reaccionar_a_entorno("lluvia")
print(f"Robot: '¡Qué buen paseo! {reserva['mascota']} se ha portado genial. Volvemos a casa.'")

# 4. Devolución y Desbloqueo de Seguridad
print("\n--- [🏠 Devolución y Seguridad] ---")
print(f"He llegado al domicilio de {reserva['mascota']}. El arnés permanece BLOQUEADO por seguridad.")

# NUEVA LÓGICA DE DESBLOQUEO AL LLEGAR
cod_final = input(f"Dueño, por seguridad ingresa el código en tu App para desbloquear el arnés ({reserva['codigo_entrega']}): ")
if cod_final == reserva['codigo_entrega']:
    print(f"¡Código correcto! Arnés liberado. Ha sido un placer cuidar de {reserva['mascota']}.")
    input("Deslice el botón en la app para confirmar recepción (Enter)")
    print("Robot: '¡Adiós! Espero que nos veamos pronto para otra aventura'.")
else:
    print("Código incorrecto. No puedo liberar el arnés. Llamando a soporte técnico por seguridad...")
    exit()

# 5. Post-Servicio Robot
print("\n--- [🔧 Mantenimiento Central] ---")
print(f"Vaciando depósito de forma ecológica: {robot.deposito_bolsas} desechos procesados.")
robot.bateria = 100
print("Estoy cargado, limpio y muy feliz por el trabajo de hoy. ¡Listo para el siguiente!")

# 6. Post-Servicio Humano (Pago y Feedback)
print("\n--- [💳 Pago y Calificación] ---")
print(f"Total a pagar por este maravilloso paseo: {reserva['total']}€")
metodo_pago = input("Seleccione método de pago: ")
propina = input("¿Deseas dejar una pequeña propina para mi mantenimiento? ")
print("Procesando el pago... ¡Todo perfecto!")

# Calificación
satis = input("Del 1 al 10, ¿te ha gustado mi compañía hoy?: ")
puntual = input("¿Qué tal me porté con la puntualidad? (1-10): ")
sugiere = input("¿Tienes alguna sugerencia para que mejore? (si/no): ").lower()
if sugiere == "si":
    input("Cuéntame, te escucho con atención: ")

print(f"\n{reserva['dueño']}, mil gracias por confiar en BotPro. {reserva['mascota']} y yo te deseamos un día fantástico.")
```
