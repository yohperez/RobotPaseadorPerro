```python
# --- TRADUCCIÓN DE PSEUDOCÓDIGO A PYTHON REAL ---

import random # Para simular eventos como charcos o necesidades

def simular_robot_paseador():
    # 1. Definición de variables (Subtarea A)
    bateria_actual = 100
    tiempo_transcurrido = 0
    tiempo_max_paseo = 10 # Paseo de 10 minutos para la prueba
    umbral_bateria_critica = 15
    umbral_tiempo_regreso = 3 # Empezar a volver 3 minutos antes
    bolsitas_recogidas = 0

    # Verificación inicial
    if bateria_actual < umbral_bateria_critica:
        print("Robot: Batería demasiado baja para salir.")
        return

    print("Robot: Perro enganchado. ¡Iniciando paseo!")
    print("Robot: Planificando ruta hacia el parque...")

    # 2. Navegación Activa (Subtarea B - Bucle principal)
    # Mientras el tiempo no se agote y la batería sea suficiente:
    while tiempo_transcurrido < tiempo_max_paseo and bateria_actual > umbral_bateria_critica:

        tiempo_transcurrido += 1
        bateria_actual -= 2 # El robot gasta energía
        print(f"\n[Minuto {tiempo_transcurrido}] Paseando por el parque...")

        # 3. Gestión de Incidentes (Subtarea C)

        # Simular detección de charco (20% de probabilidad)
        if random.random() < 0.2:
            print("Robot: ¡Charco detectado! Desviando ruta para no mojar las patas.")

        # Simular necesidad del perro (15% de probabilidad)
        if random.random() < 0.15:
            print("Robot: El perro ha hecho una necesidad. Iniciando recolección...")
            bolsitas_recogidas += 1
            print(f"Robot: Bolsita número {bolsitas_recogidas} almacenada.")

        # 4. Decisión de regreso (Subtarea D)
        tiempo_restante = tiempo_max_paseo - tiempo_transcurrido

        if tiempo_restante <= umbral_tiempo_regreso:
            print(">>> Robot: Tiempo casi agotado. Iniciando retorno a casa.")
            break # Salimos del bucle para ir a la sección de llegada

    # 5. Cierre del Paseo
    print("\n---------------------------------------")
    print("Robot: ¡Hemos llegado a casa!")
    print(f"RESUMEN: {bolsitas_recogidas} necesidades recogidas.")
    print(f"BATERÍA FINAL: {bateria_actual}%")
    print("Robot: Desenganchando correa. Notificación enviada al dueño.")

# Ejecutar la simulación
simular_robot_paseador()
```

```python
class RobotPaseador:
    def __init__(self, nombre):
        # Configuramos las características iniciales del robot
        self.nombre = nombre
        self.bateria = 100  # Porcentaje de energía
        self.bolsas = 10    # Cantidad de bolsas disponibles
        self.deposito_desechos = 0 # Contador de bolsas usadas
        self.ubicacion = "Central"

    def verificar_estado(self):
        # Paso 1: Revisar si podemos trabajar
        print(f"--- Chequeo de sistema para {self.nombre} ---")
        if self.bateria < 50:
            print("Batería baja. Cargando hasta el 100%...")
            self.bateria = 100
        else:
            print("Batería lista para la acción.")

    def gestionar_obstaculo(self, tipo):
        # Lógica para evitar charcos o lluvia
        if tipo == "charco":
            print("¡Charco detectado! Calculando ruta lateral para evitarlo.")
        elif tipo == "lluvia":
            print("Está lloviendo. Buscando refugio seco para el perrito...")
            print("Reanudando paseo tras la lluvia.")

    def recoger_desechos(self):
        # El proceso de limpieza que describiste
        print("Evento: El perro ha defecado.")
        if self.bolsas > 0:
            print("1. Sacando dispensador.")
            print("2. Tomando bolsa y recogiendo desechos.")
            print("3. Cerrando bolsa y guardando en depósito.")
            self.bolsas -= 1
            self.deposito_desechos += 1
        else:
            print("¡Error! No quedan bolsas. Informando a la central.")

    def realizar_paseo(self, perro, duracion):
        # El flujo principal del servicio
        print(f"\nIniciando servicio para: {perro}")
        print(f"Llegando al domicilio. ¡Hola, humano! Soy {self.nombre}.")
        print(f"Perro asegurado en el arnés. Iniciando ruta por la vereda.")

        # Simulamos eventos durante el paseo
        self.gestionar_obstaculo("charco")
        self.recoger_desechos()

        print(f"Paseo completado tras {duracion} minutos.")
        print("Devolviendo al perro. Por favor, deslice el botón en la app.")
        self.finalizar_dia()

    def finalizar_dia(self):
        # Regreso a la base
        self.ubicacion = "Central"
        print(f"Vaciando depósito ({self.deposito_desechos} bolsas) y recargando...")
        self.deposito_desechos = 0
        self.bateria = 100
        print("Robot listo para el siguiente turno.")

# --- EJECUCIÓN DEL EJEMPLO ---
mi_robot = RobotPaseador("RoboCan 3000")
mi_robot.verificar_estado()
mi_robot.realizar_paseo("Toby (Golden Retriever)", 30)
```

```python
import time
import random
from datetime import datetime

# -------------------------------------------------
# 1. CONFIGURACIÓN Y CLASES
# -------------------------------------------------
class RobotPaseador:
    def __init__(self):
        self.bateria = 100                     # %
        self.deposito_desechos = 0             # bolsas recogidas
        self.ubicacion = "Central"
        self.en_servicio = False
        self.botella_agua = 100                # % de agua para limpiar pis
        self.zona_libre_correa = False         # ¿entra en zona donde se puede suelta?
        self.horario_valido = True             # ¿reserva dentro del horario permitido?
        self.parque_ocupado = False            # Simula si ya hay 3 perros en el parque

    # -------------------------------------------------
    # 2. PREPARACIÓN Y VERIFICACIONES DE NORMATIVA
    # -------------------------------------------------
    def verificar_preparacion(self):
        print("\n[Robot] Verificando sistemas y normativa...")
        # ---- Batería ----
        if self.bateria < 50:
            print("❌ Batería insuficiente (<50%). Cargando...")
            self.bateria = 100
            print("✅ Batería cargada al 100%")
        else:
            print(f"✅ Batería al {self.bateria}%")

        # ---- Correa obligatoria (norma 1) ----
        if not self.confirmar_correa():
            print("❌ No se ha colocado la correa. No se puede iniciar el paseo.")
            return False

        # ---- Vacunación y chip (norma 4) ----
        if not self.confirmar_vacunacion_chip():
            print("❌ El perro no cumple con vacunación/chip. Servicio cancelado.")
            return False

        # ---- Horario del parque (norma 5) ----
        if not self.horario_valido:
            print("❌ La reserva está fuera del horario permitido del parque.")
            return False

        # ---- Ocupación del parque (norma 6) ----
        if self.parque_ocupado:
            print("❌ El parque ya tiene el número máximo de perros. Intente más tarde.")
            return False

        print("✅ Todas las verificaciones de normativa aprobadas.")
        return True

    def confirmar_correa(self):
        # Simulamos la confirmación del humano
        resp = input("¿La correa está puesta y ajustada (sí/no)? ").strip().lower()
        return resp in ("sí", "si", "s")

    def confirmar_vacunacion_chip(self):
        resp = input("¿El perro está vacunado y tiene chip activo (sí/no)? ").strip().lower()
        return resp in ("sí", "si", "s")

    # -------------------------------------------------
    # 3. GESTIÓN DE INCIDENTES (normas 2,3,7)
    # -------------------------------------------------
    def gestionar_incidente_desecho(self):
        print("\n💩 El perro ha defecado.")
        print("1️⃣ Sacando dispensador de bolsas...")
        print("2️⃣ Tomando bolsa biodegradable.")
        print("3️⃣ Inclinándose y recogiendo desechos.")
        print("4️⃣ Cerrando bolsa herméticamente.")
        print("5️⃣ Guardando en el depósito interno.")
        self.deposito_desechos += 1

    def gestionar_lluvia(self):
        print("\n🌧️ ¡Está lloviendo! Buscando refugio seco para el perro...")
        time.sleep(2)
        print("☀️ Ha parado de llover. Reanudando servicio.")

    def advertir_alimentar(self):
        print("\n⚠️ No está permitido alimentar al perro con comida humana en el parque.")
        print("🔄 Se ignora la acción y se continúa con la ruta.")

    # -------------------------------------------------
    # 4. RUTA Y ZONAS (norma 1 y 2)
    # -------------------------------------------------
    def comprobar_zona_libre_correa(self, zona):
        """
        Simula la detección de una zona habilitada para suelta.
        En la vida real se usaría un mapa GIS del parque.
        """
        zonas_habilitadas = ["Zona Libre Correa 1", "Zona Libre Correa 2"]
        self.zona_libre_correa = zona in zonas_habilitadas
        if self.zona_libre_correa:
            print(f"✅ Zona '{zona}' permite suelta de correa.")
        else:
            print(f"🔒 Zona '{zona}' requiere correa obligatoria.")

# -------------------------------------------------
# 5. INTERFAZ DE RESERVA (APP)
# -------------------------------------------------
print("--- APP ROBOT PASEADOR - MÁDRO (Madrid) ---")
input("Presione Enter para iniciar sesión...")

nombre_mascota = input("Nombre de la mascota: ").strip()
fecha_hora = input("Fecha y hora del paseo (ej: 2026-03-05 16:00): ").strip()
talla = input("Tamaño del perro (Grande/Mediano/Chico): ").strip()
tipo_servicio = input("Tipo de servicio (Tiempo/Distancia): ").strip()
domicilio = input("Dirección del domicilio: ").strip()

# -------------------------------------------------
# 6. VALIDACIÓN DE HORARIO (norma 5)
# -------------------------------------------------
# La normativa de Madrid permite paseos en parques entre 08:00 y 22:00.
# Verificamos que la hora solicitada esté dentro del rango.
def validar_horario(fecha_hora_str):
    try:
        dt = datetime.strptime(fecha_hora_str, "%Y-%m-%d %H:%M")
        hora = dt.hour
        return 8 <= hora < 22   # 08:00‑22:00
    except Exception:
        return False

robot = RobotPaseador()
robot.horario_valido = validar_horario(fecha_hora)

# -------------------------------------------------
# 7. COMPROBACIÓN DE OCUPACIÓN (norma 6)
# -------------------------------------------------
# Simulamos que el parque está libre (False). En un entorno real
# habría una base de datos con la cantidad de perros activos.
robot.parque_ocupado = False   # Cambiar a True para probar la restricción

# -------------------------------------------------
# 8. PREPARACIÓN DEL ROBOT
# -------------------------------------------------
if not robot.verificar_preparacion():
    print("\n❎ Servicio cancelado por incumplir normativa.")
    exit()

print(f"\n🤖 Robot asignado. Calculando ruta hacia {domicilio}...")
time.sleep(1)

# -------------------------------------------------
# 9. INICIO DEL SERVICIO
# -------------------------------------------------
print(f"\n[Robot] Llegando al domicilio: {domicilio}")
print("🔔 *Ding Dong* (toca el timbre)")
print(f"🤖: 'Hola, soy el robot encargado del paseo de {nombre_mascota}.'")
input(f"Coloque la correa a {nombre_mascota} en el arnés y presione Enter para confirmar...")

print("\n🚀 Iniciando paseo...")
robot.en_servicio = True
minutos_paseo = 7   # Simulación de 7 ciclos (aprox. 7 min)

# -------------------------------------------------
# 10. BUCLE DE PASEO (aplicando normas 1,2,3,7)
# -------------------------------------------------
for minuto in range(1, minutos_paseo + 1):
    print(f"\n>> Minuto {minuto}: Paseando por la vereda...")
    robot.bateria -= 2

    # Simulamos que la ruta pasa por distintas zonas
    zona_actual = random.choice(["Vereda principal", "Zona Libre Correa 1", "Vereda lateral"])
    robot.comprobar_zona_libre_correa(zona_actual)

    # Evento aleatorio
    evento = random.randint(1, 12)

    if evento == 1:          # Charco
        print("💧 Charco detectado. Recalculando para evitarlo.")
    elif evento == 2:        # Necesidad del perro
        robot.gestionar_incidente_desecho()
    elif evento == 3:        # Lluvia
        robot.gestionar_lluvia()
    elif evento == 4:        # Encuentro con otro perro
        print("🐕 Otro perro se acerca. Manteniendo distancia y ruta planificada.")
    elif evento == 5:        # El perro hace pis
        print("💧 El perro ha hecho pis. Usando botella de agua para limpiar la vereda.")
        robot.botella_agua -= 10
    elif evento == 6:        # Intento de alimentar al perro (norma 7)
        robot.advertir_alimentar()
    elif evento == 7:        # Zona libre de correa, pero el humano decide soltar
        if robot.zona_libre_correa:
            print("🦮 El perro está suelto en zona libre de correa. Vigilancia activa.")
        else:
            print("⚠️ No está permitido soltar al perro fuera de zona libre de correa.")
    else:
        print("✅ Paseo sin incidentes.")

    # Control de batería crítica
    if robot.bateria <= 15:
        print("\n⚡ Batería crítica. Iniciando regreso inmediato a la casa.")
        break

# -------------------------------------------------
# 11. DEVOLUCIÓN AL DOMICILIO
# -------------------------------------------------
print(f"\n[Robot] Regresando al domicilio {domicilio} a tiempo.")
input("Humano: Deslice el botón en la app para confirmar recepción y retirar al perro...")
print(f"🤖: '¡Fue un placer pasear con {nombre_mascota}! Hasta la próxima.'")

# -------------------------------------------------
# 12. POST‑SERVICIO (robot)
# -------------------------------------------------
print("\n--- TAREAS POST‑SERVICIO (ROBOT) ---")
print("1️⃣ Volviendo a la central...")
print(f"2️⃣ Vaciando depósito: {robot.deposito_desechos} bolsas descartadas.")
robot.bateria = 100
print("3️⃣ Recargando batería... [|||||] 100%")
print("✅ Robot listo para el siguiente servicio.\n")

# -------------------------------------------------
# 13. PAGO Y CALIFICACIÓN (humano)
# -------------------------------------------------
print("--- PAGO Y ENCUESTA DE SATISFACCIÓN ---")
metodo_pago = input("Seleccione método de pago (Tarjeta/PayPal): ").strip()
propina = input("¿Desea agregar propina? (Sí/No): ").strip().lower()
print(f"✅ Pago confirmado con {metodo_pago}. ¡Gracias por su confianza!\n")

print("📝 ENCUESTA")
p1 = input("1️⃣ ¿Qué tan satisfecho estás con el servicio? ")
p2 = input("2️⃣ ¿Cómo está el perro luego del paseo? ")
p3 = input("3️⃣ ¿Mejorarías algo del servicio? ")

print("\n✨ ¡Gracias por usar Robot Paseador en Madrid! ✨")
```
